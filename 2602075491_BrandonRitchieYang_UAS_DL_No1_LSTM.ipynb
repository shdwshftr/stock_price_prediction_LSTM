Function create_dataset untuk mempersiapkan data untuk di training di model LSTM. Parameternya adalah dataset yang akan digunakan dan step nya. dataset akan di loop lalu feature akan di buat berdasarkan jumlah step, jika step adalah 5 maka feautre yang akan digunakan untuk di training adalah 5 data dan targetnya adalah data berikutnya yaitu data ke 6.

```python
def create_dataset(dataset, step):
  X, y = [], []
  for i in range(len(dataset) - step):
    feature = dataset[i:i + step]
    target = dataset[i + step:i + step + 1]
    X.append(feature)
    y.append(target)
  return np.array(X), np.array(y)
```

Membuat fitur data menggunakan function create_dataset tadi dengan time step 5 untuk kedua data.

```python
step = 5
x_train_g, y_train_g = create_dataset(train_price_g, step = step)
x_val_g, y_val_g = create_dataset(val_price_g, step = step)
x_test_g, y_test_g = create_dataset(test_price_g, step = step)
print(x_train_g.shape, y_train_g.shape)
print(x_val_g.shape, y_val_g.shape)
print(x_test_g.shape, y_test_g.shape)
```

Membuat base model LSTM menggunakan 50 units dengan activation function relu dan input shape berupa step yang sudah di tentukan (5, 1). Lalu ada 1 fully connected layer dengan 1 unit setelahnya untuk output. Model di compile menggunakan optimizer adam dan loss function mean squared error.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import Callback

model = Sequential()
model.add(LSTM(units = 50, activation = 'relu', input_shape = (step, 1)))
model.add(Dense(units = 1))
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.summary()
```

Model di train menggunakan 100 epoch dan batch size sebesar 8 dan menggunakan validation data.

```python
history_g = model.fit(x_train_g, y_train_g, epochs = 100, batch_size = 8, validation_data = (x_val_g, y_val_g))
```

```python
history_i = model.fit(x_train_i, y_train_i, epochs = 100, batch_size = 8, validation_data = (x_val_i, y_val_i))
```

Trainig Loss dan Val Loss untuk saham INTC kurang baik karena keduanya mengalami stuck di area yang sama dan tidak mengalami penurunan. Ada spiking yang terjadi di Val loss dan Val loss lebih rendah dari training loss, ini bisa menandakan underfitting.

```python
plt.plot(history_i.history['loss'], label='Train Loss')
plt.plot(history_i.history['val_loss'], label='Val Loss')
plt.title('INTC')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

predict data testing menggunakan model yang sudah dibuat.

```python
prediction_g = model.predict(x_test_g)
prediction_i = model.predict(x_test_i)
```
